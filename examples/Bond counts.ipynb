```python
import pickle
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

atom_types = [1,2,8,6,10,26,7,14,12,16]
```

```python
bonds = pickle.load(open("../etc/bonds.p", "rb"))
bond_counts = {key: Counter(bonds[key]) for key in atom_types}
```

```python
for i in atom_types:
    plt.figure()
    plt.title("Atomic number: %i"%i)
    plt.bar(left=bond_counts[i].keys(), height=bond_counts[i].values())
```
